# **Landfill Environmental Justice Data Story**

This notebook performs an analysis on Landfill plumes from EMIT data for Dallas, Texas and Stockton, California. The goal is to explore and visualize the data, understand its distribution, and identify any overlaps between the plumes and socioeconomic status using census data from the CDC.

```python
import geopandas as gpd
import numpy as np
import rasterio
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature
from rasterio.plot import show
import matplotlib.gridspec as gridspec
import matplotlib.patheffects as path_effects
```

# **Read and Plot Dallas, TX Plumes with a scale of 1 to 1500 ppm m**

This section contains code to read or load the data. It is a crucial step as it brings the data into the for analysis.

```python
# File paths
file_paths = [
    'new-landfill/dallas_1.tiff',
    'new-landfill/dallas_2.tiff',
    'new-landfill/dallas_3.tiff'
]

# Set the desired min and max values for rescaling
vmin, vmax = 1, 1500

# Create a figure with subplots
fig = plt.figure(figsize=(20, 12))
gs = gridspec.GridSpec(2, 3, height_ratios=[10, 1], hspace=0.3)

axes = [fig.add_subplot(gs[0, i]) for i in range(3)]

# Plot each TIFF file with 'inferno' colormap
for ax, file_path in zip(axes, file_paths):
    with rasterio.open(file_path) as src:
        img = show(src, ax=ax, cmap='inferno', vmin=vmin, vmax=vmax, title=file_path.split('/')[-1])

# Create a colorbar
cbar_ax = fig.add_subplot(gs[1, :])
cbar = plt.colorbar(img.get_images()[0], cax=cbar_ax, orientation='horizontal')
cbar.set_label('Methane (ppm m)')

plt.show()
```

This section contains code that extracts the central coordinates of each of the plumes to annotate on our SVI map.

```python
# Function to calculate the center coordinates of a raster
def get_center_coords(file_path):
    with rasterio.open(file_path) as src:
        bounds = src.bounds
        center_x = (bounds.left + bounds.right) / 2
        center_y = (bounds.bottom + bounds.top) / 2
    return center_x, center_y

# Extract and print the central coordinates for each TIFF
for file_path in file_paths:
    center_coords = get_center_coords(file_path)
    print(f'Central coordinates of {file_path.split("/")[-1]}: {center_coords}')
```

# **CDC SVI data from 2020 for Texas**

This section contains code that reads in the census data and plots the overall Texas SVI and the counties of interest with Dallas and Fort Worth annotated. We also annotated where the plumes are located.

```python
texas_svi = 'new-landfill/SVI2020_TEXAS_tract.gdb'
texas_svi = gpd.read_file(texas_svi)
```

```python
state_name = 'Texas'
column_to_plot = 'RPL_THEMES'

# Filter the DataFrame for Texas
texas_svi = texas_svi[texas_svi['STATE'] == state_name]

# Remove invalid values
texas_svi = texas_svi[texas_svi[column_to_plot] != -999]

# Plot the overall SVI for Texas
plt.figure(figsize=(10, 6))
texas_svi.plot(column=column_to_plot, legend=True, cmap='rainbow', vmin=0, vmax=1)
plt.title('SVI 2020 - Texas (Overall SVI)')
plt.axis('off')
plt.show()
```

```python
# Central coordinates of the plume TIFFs
plume_coords = [
    (-96.96554886988591, 33.01865445063364),
    (-96.76627841869168, 32.67379456775059),
    (-96.75407818698592, 32.546098809230216)
]

# Filter for Dallas, Tarrant, and surrounding counties
counties = texas_svi[(texas_svi['COUNTY'] == 'Dallas') | (texas_svi['COUNTY'] == 'Tarrant') |
                     (texas_svi['COUNTY'] == 'Collin') | (texas_svi['COUNTY'] == 'Denton') |
                     (texas_svi['COUNTY'] == 'Johnson') | (texas_svi['COUNTY'] == 'Ellis')]

# Plot the SVI for Dallas and Tarrant counties
fig, ax = plt.subplots(figsize=(10, 6))
counties.plot(column=column_to_plot, legend=True, cmap='rainbow', vmin=0, vmax=1, ax=ax)

# Annotate Dallas and Fort Worth locations
# Approximate coordinates for the annotations (you might need to adjust these)
dallas_coords = (-96.7969, 32.7767)
fort_worth_coords = (-97.3308, 32.7555)

# Add annotations for Dallas and Fort Worth
text = ax.annotate('Dallas', xy=dallas_coords, xytext=(dallas_coords[0]+0.1, dallas_coords[1]+0.1),
                   arrowprops=dict(facecolor='black', arrowstyle='->'), color='black')
text.set_path_effects([path_effects.Stroke(linewidth=3, foreground='white'), path_effects.Normal()])
text = ax.annotate('Fort Worth', xy=fort_worth_coords, xytext=(fort_worth_coords[0]+0.1, fort_worth_coords[1]+0.1),
                   arrowprops=dict(facecolor='black', arrowstyle='->'), color='black')
text.set_path_effects([path_effects.Stroke(linewidth=3, foreground='white'), path_effects.Normal()])

# Annotate the central coordinates of the plume TIFFs
for i, coords in enumerate(plume_coords):
    ax.scatter(*coords, edgecolor='black', color='magenta', s=100, label=f'Plume {i+1}' if i == 0 else "")
    text = ax.annotate(f'Plume {i+1}', xy=coords, xytext=(coords[0]+0.05, coords[1]-0.05), color='black')
    text.set_path_effects([path_effects.Stroke(linewidth=3, foreground='white'), path_effects.Normal()])

plt.title('SVI 2020 - Dallas, Tarrant, and Surrounding Counties (Texas)')
plt.axis('off')
plt.show()
```

# **CDC SVI data from 2020 for California**

This section contains code that reads in the census data and plots the overall Texas SVI and the counties of interest with Dallas and Fort Worth annotated. We also annotated where the plumes are located.

```python
california_svi = 'new-landfill/SVI2020_CALIFORNIA_tract.gdb'
california_svi = gpd.read_file(california_svi)
```

```python
state_name = 'California'
column_to_plot = 'RPL_THEMES'

california_svi = california_svi[california_svi['STATE'] == state_name]

california_svi = california_svi[california_svi[column_to_plot] != -999]

plt.figure(figsize=(10, 6))
california_svi.plot(column=column_to_plot, legend=True, cmap='rainbow', vmin=0, vmax=1)
plt.title('SVI 2020 - Texas (Overall SVI)')
plt.axis('off')
plt.show()
```

```python
# Central coordinates of the plume TIFF
central_coords = (-121.17487531803246, 37.87624448335031)

# Stockton coordinates
stockton_coords = (-121.2908, 37.9577)

# Filter for San Joaquin County
san_joaquin_svi = california_svi[california_svi['COUNTY'] == 'San Joaquin']

# Plot the SVI for San Joaquin County
fig, ax = plt.subplots(figsize=(10, 6))
san_joaquin_svi.plot(column=column_to_plot, legend=True, cmap='rainbow', vmin=0, vmax=1, ax=ax)

# Annotate Stockton location
ax.annotate('Stockton', xy=stockton_coords, xytext=(stockton_coords[0]+0.1, stockton_coords[1]+0.1),
            arrowprops=dict(facecolor='black', arrowstyle='->'), color='black',
            path_effects=[path_effects.Stroke(linewidth=3, foreground='white'), path_effects.Normal()])

# Annotate the central coordinates of the plume TIFF
ax.scatter(*central_coords, color='magenta', s=100, label='Plume', edgecolor='black')
ax.annotate('Plume', xy=central_coords, xytext=(central_coords[0]+0.05, central_coords[1]-0.05), color='black',
            path_effects=[path_effects.Stroke(linewidth=3, foreground='white'), path_effects.Normal()])

plt.title('SVI 2020 - San Joaquin County (California)')
plt.axis('off')
plt.show()
```
